```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/thedhruvish/dataset-machine-learning/refs/heads/main/ml-algorithms-project/College.csv")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_numeric = df.drop(columns=['Private'])
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_numeric)
```

```python
dbscan = DBSCAN(eps=2, min_samples=5)
clusters = dbscan.fit_predict(scaled_data)
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
```

```python
vis_df = pd.DataFrame(pca_result, columns=['PCA1', 'PCA2'])
vis_df['Cluster'] = clusters
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=vis_df, x='PCA1', y='PCA2', hue='Cluster', palette='tab10', s=50)
plt.title("DBSCAN Clusters Visualized in 2D PCA Space")
plt.legend(title='Cluster')
plt.show()
```
