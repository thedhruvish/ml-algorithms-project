```python


# Load the data
df = pd.read_csv("musk.csv")

# Drop non-numeric or identifier columns if necessary
X = df.select_dtypes(include=["number"])

# Optional: Drop any ID or label columns if present
# X = X.drop(columns=["ID", "Label"])

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Run KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Visualize with PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='Set1')
plt.title("KMeans Clustering of Musk Dataset")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.colorbar(label="Cluster")
plt.show()
```
